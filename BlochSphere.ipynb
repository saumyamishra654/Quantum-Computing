```python
from qiskit import * ### Importing the library to our notebook
%matplotlib inline
q = QuantumRegister(1, 'q') ### (No. of qubits, name of the qubit(s))
qc=QuantumCircuit(q)
qc.draw(output='mpl') ### Drawing the quantum circuit
```

```python
# from qiskit_aer import AerSimulator
# from qiskit_aer import AerProvider
prov = AerProvider()
sv_simulator=prov.get_backend('statevector_simulator') ### To generate the statevect
result=sv_simulator.run(qc).result()
random_number=result.get_statevector()
random_number
```

```python
from qiskit.quantum_info import Statevector
sv = Statevector.from_label('0')
sv=sv.evolve(qc)
sv.draw('latex') ### Printing the statevector
```

```python
from qiskit.visualization import plot_bloch_multivector
plot_bloch_multivector(sv) ### Drawing the bloch-sphere
```

```python
### Quantum true-random number generator (qtrng)
q = QuantumRegister(1, 'q')
c = ClassicalRegister(1, 'c')
qc1 = QuantumCircuit(q,c)
qc1.h(q)
qc1.measure(q,c)
qc1.draw(output='mpl')
```

```python
result=sv_simulator.run(qc1).result()
random_number=result.get_statevector()
random_number
```

```python
from qiskit.visualization import plot_histogram
simulator = prov.get_backend('qasm_simulator')

simulator = simulator.run(qc1).result()
counts = simulator.get_counts(qc1)
plot_histogram(counts)
```
